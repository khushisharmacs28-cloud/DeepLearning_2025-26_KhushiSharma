```python
def add_expense(expenses):
    amount = float(input("Enter the expense amount: "))
    category = input("Enter the category: ")

    if category in expenses:
        expenses[category].append(amount)
    else:
        expenses[category] = [amount]

def choose_category():
    print("Available categories:")
    for category in expenses:
        print("-", category)
    return input("Choose a category: ")

def calculate_total_expenses(expenses):
    total_expenses = {}
    for category, amounts in expenses.items():
        total_expenses[category] = sum(amounts)
    return total_expenses

def display_summary(total_expenses):
    print("Expense Summary:")
    for category, total in total_expenses.items():
        print(f"{category}: ${total:.2f}")

def main():
    expenses = {}
    while True:
        print("\nExpense Tracker Menu:")
        print("1. Add New Expense")
        print("2. Display Summary")
        print("3. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_expense(expenses)
        elif choice == '2':
            total_expenses = calculate_total_expenses(expenses)
            display_summary(total_expenses)
        elif choice == '3':
            print("Exiting Expense Tracker. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
def add_task(todo_list):
    task = input("Enter the new task: ")
    todo_list.append({"task": task, "completed": False})

def mark_completed(todo_list):
    print("To-Do List:")
    for index, item in enumerate(todo_list):
        print(f"{index + 1}. {item['task']} - {'Completed' if item['completed'] else 'Pending'}")

    choice = int(input("Enter the index of the task to mark as completed: ")) - 1
    if 0 <= choice < len(todo_list):
        todo_list[choice]['completed'] = True
    else:
        print("Invalid index!")

def delete_task(todo_list):
    print("To-Do List:")
    for index, item in enumerate(todo_list):
        print(f"{index + 1}. {item['task']} - {'Completed' if item['completed'] else 'Pending'}")

    choice = int(input("Enter the index of the task to delete: ")) - 1
    if 0 <= choice < len(todo_list):
        del todo_list[choice]
    else:
        print("Invalid index!")

def display_todo_list(todo_list):
    print("To-Do List:")
    for index, item in enumerate(todo_list):
        print(f"{index + 1}. {item['task']} - {'Completed' if item['completed'] else 'Pending'}")

def main():
    todo_list = []
    while True:
        print("\nTo-Do List Manager Menu:")
        print("1. Add New Task")
        print("2. Mark Task as Completed")
        print("3. Delete Task")
        print("4. Display To-Do List")
        print("5. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            add_task(todo_list)
        elif choice == '2':
            mark_completed(todo_list)
        elif choice == '3':
            delete_task(todo_list)
        elif choice == '4':
            display_todo_list(todo_list)
        elif choice == '5':
            print("Exiting To-Do List Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
def log_food(food_log):
    food_name = input("Enter the name of the food item: ")
    calorie_count = float(input("Enter the calorie count of the food item: "))
    food_log.append({"food_name": food_name, "calorie_count": calorie_count})

def calculate_total_calories(food_log):
    total_calories = sum(item['calorie_count'] for item in food_log)
    return total_calories

def set_daily_goal():
    return float(input("Enter your daily calorie goal: "))

def provide_feedback(total_calories, daily_goal):
    if total_calories <= daily_goal:
        print("Congratulations! You've reached your daily calorie goal.")
    else:
        print(f"You've exceeded your daily calorie goal. Total calories consumed: {total_calories:.2f}")

def main():
    food_log = []
    daily_goal = set_daily_goal()

    while True:
        print("\nCalorie Counter Menu:")
        print("1. Log Food Item Consumed")
        print("2. Calculate Total Calories Consumed")
        print("3. Provide Feedback on Progress")
        print("4. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            log_food(food_log)
        elif choice == '2':
            total_calories = calculate_total_calories(food_log)
            print(f"Total calories consumed: {total_calories:.2f}")
        elif choice == '3':
            total_calories = calculate_total_calories(food_log)
            provide_feedback(total_calories, daily_goal)
        elif choice == '4':
            print("Exiting Calorie Counter. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
# Python Code for Polymorphism

# Base class
class Animal:
    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")

# Derived class 1
class Dog(Animal):
    def speak(self):
        return "Woof!"

# Derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow!"

# Derived class 3
class Bird(Animal):
    def speak(self):
        return "Chirp!"

# Function demonstrating polymorphism
def animal_sound(animal):
    print(animal.speak())

# Instances of derived classes
dog = Dog()
cat = Cat()
bird = Bird()

# Demonstrating method overriding
print("Method Overriding:")
animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!
animal_sound(bird) # Output: Chirp!

# Operator Overloading
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

# Creating vectors
v1 = Vector(2, 3)
v2 = Vector(4, 5)

# Adding vectors using overloaded + operator
v3 = v1 + v2
print("\nOperator Overloading:")
print(v3)  # Output: Vector(6, 8)

# Polymorphism in Collections
animals = [dog, cat, bird]

print("\nPolymorphism in Collections:")
for animal in animals:
    print(animal.speak())
```

```python
# Example of catching different types of exceptions
# SyntaxError: Raised when there is a syntax error in the code.
# IndentationError: Raised when there is incorrect indentation in the code.
# NameError: Raised when a local or global name is not found.
# TypeError: Raised when an operation or function is applied to an object of inappropriate type.
# ValueError: Raised when a built-in operation or function receives an argument that has the right type but an inappropriate value.
# ZeroDivisionError: Raised when the second operand of a division or modulo operation is zero.
# IndexError: Raised when a sequence subscript is out of range.
# KeyError: Raised when a dictionary key is not found.
# FileNotFoundError: Raised when a file or directory is requested but cannot be found.
# ImportError: Raised when an import statement fails to find the module definition.


def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: Division by zero!")
    except TypeError:
        print("Error: Unsupported operand type!")
    except Exception as e:
        print("An unexpected error occurred:", e)
    else:
        print("Result:", result)
    finally:
        print("Execution completed.")

# Test cases
divide(10, 2)  # Result: 5.0
divide(10, 0)  # Error: Division by zero!
divide("10", 2)  # Error: Unsupported operand type!
divide(10, "2")  # An unexpected error occurred: unsupported operand type(s) for /: 'int' and 'str'
```

```python
# Python Code for Regular Expressions
# The re module in Python provides support for working with regular expressions. You can use it to search, match, and manipulate strings based on patterns.


# Metacharacters:

# .: Matches any character except a newline. findall returns all characters except newlines.
# ^: Matches the start of the string. Finds 'a' only at the start.
# $: Matches the end of the string. Finds 'c' only at the end.
# *: Matches 0 or more repetitions. Finds 'a' followed by any number of 'b's.
# +: Matches 1 or more repetitions. Finds 'a' followed by one or more 'b's.
# ?: Matches 0 or 1 repetition. Finds 'a' followed by zero or one 'b'.
# {}: Matches a specific number of repetitions. Finds 'a' followed by 2 to 3 'b's.
# []: Matches any character in the brackets. Finds any character 'a', 'b', or 'c'.
# |: Matches either pattern. Finds either 'a' or 'b'.
# (): Groups patterns. Finds repeated 'ab' groups.

# Special Sequences:

# \d: Matches any digit.
# \D: Matches any non-digit.
# \s: Matches any whitespace.
# \S: Matches any non-whitespace.
# \w: Matches any alphanumeric character.
# \W: Matches any non-alphanumeric character.


import re

# Metacharacters

# . - Matches any character except a newline
pattern_dot = re.compile(r".")
print(pattern_dot.findall("a\nb c"))  # ['a', 'b', ' ', 'c']

# ^ - Matches the start of the string
pattern_caret = re.compile(r"^a")
print(pattern_caret.findall("abc ab"))  # ['a']

# $ - Matches the end of the string
pattern_dollar = re.compile(r"c$")
print(pattern_dollar.findall("abc ab"))  # ['c']

# * - Matches 0 or more repetitions of the preceding element
pattern_star = re.compile(r"ab*")
print(pattern_star.findall("abbb aab"))  # ['abbb', 'a', 'ab']

# + - Matches 1 or more repetitions of the preceding element
pattern_plus = re.compile(r"ab+")
print(pattern_plus.findall("abbb aab"))  # ['abbb', 'ab']

# ? - Matches 0 or 1 repetition of the preceding element
pattern_question = re.compile(r"ab?")
print(pattern_question.findall("abbb aab"))  # ['ab', 'a', 'ab']

# {} - Matches exactly m times if {m}, at least m times if {m,}, and from m to n times if {m,n}
pattern_braces = re.compile(r"ab{2,3}")
print(pattern_braces.findall("ab abbb abb abbbb"))  # ['abb', 'abbb', 'abb']

# [] - Matches any single character within the brackets
pattern_brackets = re.compile(r"[a-c]")
print(pattern_brackets.findall("abc d efg"))  # ['a', 'b', 'c']

# | - Matches either the expression before or the expression after the |
pattern_pipe = re.compile(r"a|b")
print(pattern_pipe.findall("abc"))  # ['a', 'b']

# () - Groups patterns together and captures the matched text
pattern_parentheses = re.compile(r"(ab)+")
print(pattern_parentheses.findall("ab ab ab"))  # ['ab', 'ab', 'ab']

### Special Sequences

# \d - Matches any digit (equivalent to [0-9])
pattern_digit = re.compile(r"\d")
print(pattern_digit.findall("a1b2c3"))  # ['1', '2', '3']

# \D - Matches any non-digit (equivalent to [^0-9])
pattern_non_digit = re.compile(r"\D")
print(pattern_non_digit.findall("a1b2c3"))  # ['a', 'b', 'c']

# \s - Matches any whitespace character (equivalent to [ \t\n\r\f\v])
pattern_whitespace = re.compile(r"\s")
print(pattern_whitespace.findall("a b\tc\n"))  # [' ', '\t', '\n']

# \S - Matches any non-whitespace character (equivalent to [^ \t\n\r\f\v])
pattern_non_whitespace = re.compile(r"\S")
print(pattern_non_whitespace.findall("a b\tc\n"))  # ['a', 'b', 'c']

# \w - Matches any alphanumeric character (equivalent to [a-zA-Z0-9_])
pattern_word = re.compile(r"\w")
print(pattern_word.findall("a_b-c!"))  # ['a', 'b', 'c']

# \W - Matches any non-alphanumeric character (equivalent to [^a-zA-Z0-9_])
pattern_non_word = re.compile(r"\W")
print(pattern_non_word.findall("a_b-c!"))  # ['-', '!']

# 1. Basic Pattern Matching
pattern = r'hello'
text = 'hello world! Hello everyone.'

# Search for the pattern in the text
match = re.search(pattern, text)
if match:
    print("Basic Pattern Matching:")
    print(f"Found '{match.group()}' at position {match.start()}")  # Output: Found 'hello' at position 0

# 2. Case Insensitive Matching
pattern = r'Hello'
match = re.search(pattern, text, re.IGNORECASE)
if match:
    print("\nCase Insensitive Matching:")
    print(f"Found '{match.group()}' at position {match.start()}")  # Output: Found 'hello' at position 0

# 3. Using Capturing Groups
pattern = r'(\d{3})-(\d{2})-(\d{4})'
text = 'My number is 123-45-6789.'

match = re.search(pattern, text)
if match:
    print("\nUsing Capturing Groups:")
    print(f"Full match: {match.group(0)}")  # Output: Full match: 123-45-6789
    print(f"Group 1: {match.group(1)}")     # Output: Group 1: 123
    print(f"Group 2: {match.group(2)}")     # Output: Group 2: 45
    print(f"Group 3: {match.group(3)}")     # Output: Group 3: 6789



# 4. Character Classes
pattern = r'[aeiou]'
text = 'The quick brown fox.'

# Find all vowels in the text
matches = re.findall(pattern, text)
print("\nCharacter Classes:")
print(f"Vowels found: {matches}")  # Output: Vowels found: ['e', 'u', 'i', 'o', 'o']

# 5. Quantifiers
pattern = r'\d+'
text = 'There are 123 apples, 456 oranges, and 789 bananas.'

# Find all sequences of digits
matches = re.findall(pattern, text)
print("\nQuantifiers:")
print(f"Numbers found: {matches}")  # Output: Numbers found: ['123', '456', '789']

# 6. Common Regex Functions

# findall: Find all matches
pattern = r'\w+'
text = 'This is a test.'
words = re.findall(pattern, text)
print("\nCommon Regex Functions:")
print(f"Words found: {words}")  # Output: Words found: ['This', 'is', 'a', 'test']

# sub: Substitute patterns in a string
pattern = r'\d+'
text = 'My numbers are 123 and 456.'
replaced_text = re.sub(pattern, '###', text)
print(f"Substitution result: {replaced_text}")  # Output: Substitution result: My numbers are ### and ###.

# split: Split a string by the occurrences of a pattern
pattern = r'\s+'
text = 'Split this text into words.'
split_text = re.split(pattern, text)
print(f"Split result: {split_text}")  # Output: Split result: ['Split', 'this', 'text', 'into', 'words.']

# 7. Anchors and Boundaries
pattern = r'^\w+'
text = 'Start at the beginning.'

# Match the start of the string
match = re.match(pattern, text)
if match:
    print("\nAnchors and Boundaries:")
    print(f"Match at start: {match.group()}")  # Output: Match at start: Start

pattern = r'\d+$'
text = 'End at the last word.'

# Match the end of the string
match = re.search(pattern, text)
if match:
    print(f"Match at end: {match.group()}")  # Output: Match at end: word

# 8. Escaping Special Characters
pattern = r'\$100\.00'
text = 'The price is $100.00.'

# Search for the price pattern
match = re.search(pattern, text)
if match:
    print("\nEscaping Special Characters:")
    print(f"Found price: {match.group()}")  # Output: Found price: $100.00


```

```python
# Python Code for Greedy vs Non-Greedy Matching

import re

# Example text
text = "<div>Some content</div><div>More content</div>"

# 1. Greedy Matching
# Greedy match attempts to match the longest possible string
greedy_pattern = r'<div>.*</div>'
greedy_match = re.search(greedy_pattern, text)
if greedy_match:
    print("Greedy Matching:")
    print(f"Matched: {greedy_match.group()}")  # Output: Matched: <div>Some content</div><div>More content</div>

# 2. Non-Greedy (Lazy) Matching
# Non-greedy match attempts to match the shortest possible string
non_greedy_pattern = r'<div>.*?</div>'
non_greedy_match = re.search(non_greedy_pattern, text)
if non_greedy_match:
    print("\nNon-Greedy (Lazy) Matching:")
    print(f"Matched: {non_greedy_match.group()}")  # Output: Matched: <div>Some content</div>

# 3. Using findall with both patterns
# findall to find all matches in the text
greedy_matches = re.findall(greedy_pattern, text)
non_greedy_matches = re.findall(non_greedy_pattern, text)

print("\nfindall with Greedy Pattern:")
print(greedy_matches)  # Output: ['<div>Some content</div><div>More content</div>']

print("\nfindall with Non-Greedy Pattern:")
print(non_greedy_matches)  # Output: ['<div>Some content</div>', '<div>More content</div>']
```

```python
# Abstract Data Types (ADTs)
# An Abstract Data Type is a theoretical concept that defines a data structure solely by its behavior from the point of view of a user. ADTs are not concerned with how the data structure is implemented, but rather what operations can be performed on it and what the behavior of these operations should be.

# Characteristics of ADTs:
# Abstraction: Focuses on what operations are performed rather than how they are performed.
# Encapsulation: Hides the implementation details from the user.
# User Interface: Defines operations that can be performed and their expected behavior.
# Examples of ADTs:
# Stack:
# Operations: push, pop, peek, isEmpty.
# Queue:
# Operations: enqueue, dequeue, front, isEmpty.
# List:
# Operations: add, remove, get, set, size.
# Tree:
# Operations: insert, delete, search, traverse.

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("pop from empty stack")

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
```

```python
# Primitive Data Types
# Primitive Data Types are the most basic data types provided by a programming language. They serve as the building blocks for data manipulation and are typically supported directly by the hardware of the computer.

# Characteristics of Primitive Data Types:
# Built-in: Provided by the language and not composed of other data types.
# Simple: Represent single values.
# Efficient: Directly supported by the machine's hardware, making them fast to operate on.
# Common Primitive Data Types:
# Integer (int): Represents whole numbers.
# Floating-point (float): Represents numbers with fractional parts.
# Character (char): Represents a single character (e.g., 'a', '1', '!').
# Boolean (bool): Represents truth values (True or False).

# Integer
x = 42
print(type(x))  # Output: <class 'int'>

# Floating-point
y = 3.14
print(type(y))  # Output: <class 'float'>

# Character (Python doesn't have a char type, uses str for single characters)
char = 'a'
print(type(char))  # Output: <class 'str'>

# Boolean
is_valid = True
print(type(is_valid))  # Output: <class 'bool'>


# Key Differences
# Level of Abstraction:

# ADTs focus on high-level design, defining data and operations abstractly.
# Primitive data types are low-level, fundamental building blocks provided by the language.
# Complexity:

# ADTs can be complex and composed of multiple elements or other types.
# Primitive data types are simple and not composed of other types.
# Implementation:

# ADTs are typically implemented by the programmer or provided by libraries.
# Primitive data types are directly supported by the language and often mapped directly to hardware operations.
```

```python
# Scenario: A To-Do List Application
# Primitive Data Types
# In our to-do list application, we'll use primitive data types to represent basic pieces of information:

# String (str): To represent the task description.
# Boolean (bool): To represent whether a task is completed.

# Primitive data types example

# Task description (string)
task_description = "Buy groceries"

# Task completion status (boolean)
is_completed = False

print("Task:", task_description)
print("Completed:", is_completed)

# Abstract Data Types
# Now, let's consider how we might manage a collection of tasks in a more structured way using an Abstract Data Type (ADT). We'll define a ToDoList ADT that allows us to add, remove, and check tasks.

class ToDoList:
    def __init__(self):
        self.tasks = []  # A list to store tasks

    def add_task(self, description):
        task = {
            "description": description,
            "completed": False
        }
        self.tasks.append(task)

    def remove_task(self, description):
        self.tasks = [task for task in self.tasks if task["description"] != description]

    def mark_task_completed(self, description):
        for task in self.tasks:
            if task["description"] == description:
                task["completed"] = True

    def get_tasks(self):
        return self.tasks

# Using the ToDoList ADT
todo_list = ToDoList()
todo_list.add_task("Buy groceries")
todo_list.add_task("Read a book")
print("Tasks before completion:")
print(todo_list.get_tasks())
todo_list.mark_task_completed("Buy groceries")
print("Tasks after completion:")
print(todo_list.get_tasks())
todo_list.remove_task("Read a book")
print("Tasks after removal:")
print(todo_list.get_tasks())


# Primitive Data Types:

# task_description and is_completed are simple values using primitive types str and bool.
# Abstract Data Types:

# ToDoList is an ADT that encapsulates the behavior of a list of tasks. It allows us to:
# Add a task: add_task(description)
# Remove a task: remove_task(description)
# Mark a task as completed: mark_task_completed(description)
# Retrieve all tasks: get_tasks()
```

```python
# List Comprehension
# Example: Squaring Numbers in a List

# Traditional For-Loop Approach
squares = []
for i in range(1, 11):
    squares.append(i ** 2)
print("Squares using traditional for-loop:", squares)

# List Comprehension Approach
squares = [i ** 2 for i in range(1, 11)]
print("Squares using list comprehension:", squares)

# More Complex Example: Filtering and Transforming
# Traditional For-Loop Approach
even_squares = []
for i in range(1, 11):
    if i % 2 == 0:
        even_squares.append(i ** 2)
print("Even squares using traditional for-loop:", even_squares)

# List Comprehension Approach
even_squares = [i ** 2 for i in range(1, 11) if i % 2 == 0]
print("Even squares using list comprehension:", even_squares)

# Explanation:
# The list comprehension `[i ** 2 for i in range(1, 11) if i % 2 == 0]` adds a conditional statement:
# - `if i % 2 == 0`: This filters the elements, only including those where the condition is `True`.

# Using Nested List Comprehensions

# Example: Creating a 3x3 Matrix

matrix = [[j for j in range(3)] for i in range(3)]
print("3x3 matrix using nested list comprehensions:")
print(matrix)

# Explanation:
# The outer list comprehension `[... for i in range(3)]` creates the rows.
# The inner list comprehension `[j for j in range(3)]` creates the columns.
# This creates a 3x3 matrix:
# [[0, 1, 2],
#  [0, 1, 2],
#  [0, 1, 2]]
```

```python
# Calculator Using GUI
import tkinter as tk

# Function to update the entry widget text
def btn_click(item):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(tk.END, current + str(item))

# Function to clear the entry widget
def btn_clear():
    entry.delete(0, tk.END)

# Function to calculate the expression in the entry widget
def btn_equal():
    expression = entry.get()
    try:
        result = str(eval(expression))  # Evaluate the expression
        entry.delete(0, tk.END)
        entry.insert(tk.END, result)
    except Exception as e:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

# Create the main window
root = tk.Tk()
root.title("Simple Calculator")
root.geometry("400x500")

# Create an entry widget for the calculator display
entry = tk.Entry(root, font=("Arial", 60), bd=10, insertwidth=5, width=20, borderwidth=10)
entry.grid(row=0, column=0, columnspan=8,padx=20, pady=20)

# Define buttons
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('.', 4, 1), ('+', 4, 2), ('=', 4, 3),
]

# Create button widgets and place them on the grid
for (text, row, col) in buttons:
    if text == '=':
        button = tk.Button(root, text=text, padx=20, pady=20, bd=8, fg="black", font=("Arial", 18),
                           command=btn_equal)
    else:
        button = tk.Button(root, text=text, padx=20, pady=20, bd=8, fg="black", font=("Arial", 18),
                           command=lambda t=text: btn_click(t))
    button.grid(row=row, column=col, sticky=tk.NSEW)

# Create clear button
button_clear = tk.Button(root, text='C', padx=20, pady=20, bd=8, fg="black", font=("Arial", 18), command=btn_clear)
button_clear.grid(row=5, column=0, columnspan=4, sticky=tk.NSEW)

# Configure row and column weights for proper resizing
for i in range(6):
    root.grid_rowconfigure(i, weight=1)
for i in range(4):
    root.grid_columnconfigure(i, weight=1)

# Run the application
root.mainloop()
```

```python
import sqlite3
from datetime import datetime

# Connect to SQLite (in memory or file-based)
conn = sqlite3.connect("calculator_logs.db")  # Saved in Colab session
cursor = conn.cursor()

# Create table
cursor.execute('''
CREATE TABLE IF NOT EXISTS logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    num1 REAL,
    num2 REAL,
    operation TEXT,
    result REAL,
    timestamp TEXT
)
''')
conn.commit()

# Calculate and Log
def calculate_and_log():
    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        op = input("Enter operation (+, -, *, /): ")

        if op == '+':
            result = num1 + num2
        elif op == '-':
            result = num1 - num2
        elif op == '*':
            result = num1 * num2
        elif op == '/':
            result = num1 / num2 if num2 != 0 else float('inf')
        else:
            print("Invalid operation!")
            return

        timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        cursor.execute("INSERT INTO logs (num1, num2, operation, result, timestamp) VALUES (?, ?, ?, ?, ?)",
                       (num1, num2, op, result, timestamp))
        conn.commit()
        print(f"Result: {result} | Logged at: {timestamp}")
    except Exception as e:
        print("Error:", e)

# View Logs
def view_logs():
    cursor.execute("SELECT * FROM logs")
    rows = cursor.fetchall()
    print("\n--- Logs ---")
    for row in rows:
        print(row)

# Update Log
def update_log():
    try:
        log_id = int(input("Enter log ID to update: "))
        new_result = float(input("Enter new result: "))
        cursor.execute("UPDATE logs SET result = ? WHERE id = ?", (new_result, log_id))
        conn.commit()
        print("Log updated.")
    except Exception as e:
        print("Error:", e)

# Delete Log
def delete_log():
    try:
        log_id = int(input("Enter log ID to delete: "))
        cursor.execute("DELETE FROM logs WHERE id = ?", (log_id,))
        conn.commit()
        print("Log deleted.")
    except Exception as e:
        print("Error:", e)

# Menu System
def menu():
    while True:
        print("\nMenu:\n1. Calculate\n2. View Logs\n3. Update Log\n4. Delete Log\n5. Exit")
        choice = input("Choose an option: ")

        if choice == '1':
            calculate_and_log()
        elif choice == '2':
            view_logs()
        elif choice == '3':
            update_log()
        elif choice == '4':
            delete_log()
        elif choice == '5':
            print("Exiting...")
            break
        else:
            print("Invalid choice!")

menu()
conn.close()
```
